# Equations of State

In physical chemistry, an equation of state (EOS) is a mathematical relationship that describes the thermodynamic behavior of a system, particularly the relationship between the state variables such as pressure (P), volume (V), and temperature (T). Equations of state are fundamental tools used to understand and predict the properties of gases, liquids, and solids under different conditions. A generic volumetric equation of state:

$$ V = f(P,T,n) =\frac{nRT}{P}$$

Read the abov line as *Volume is some function of $P$, $T$ and $n$ and they are algebraically related by the following expression.*

Equations of state are used to represent the behavior of a substance over a wide range of conditions, from idealized cases to more complex and realistic scenarios. They play a crucial role in various areas of physical chemistry, including thermodynamics, statistical mechanics, and the study of phase transitions. Equations of state (EOS) can get complicated easily. This is where tools like SymPy help us a great deal. We have have a couple more tricks to help simplify working with EOS, the most common is the use of `molar volume`. Until we start looking at the thermodynamics of chemical reactions we will always assume constant moles at n=1. To indicate that we are doing this we will replace all occurences of $V$ with $V_m$ where strictly speaking $V_m = \frac{V}{n}$ this is called the `molar volume` and it represents the volume of 1 mole of that substance. Think of it as an inverse density. Often times the $V_m$ will just be written as $V$ which can be confusing.

Some commonly used equations of state include:

## Ideal Gas Law
The simplest equation of state, which applies to ideal gases. It states that the product of pressure and volume is proportional to the product of the number of moles and the temperature.

$$PV_m = RT$$

## Van der Waals Equation of State
A more sophisticated equation that accounts for the finite size of gas molecules and the attractive forces between them. It's an improvement over the ideal gas law.

$$\left(P + \frac{a}{V_m^2}\right)(V_m - b) = nRT$$

## Redlich-Kwong Equation of State
Another improvement over the ideal gas law that includes temperature-dependent parameters and better accounts for non-ideal behavior.

$$P = \frac{RT}{V_m - b} - \frac{a}{\sqrt{T}V_m(V_m+b)}$$

## Peng-Robinson Equation of State
A widely used equation of state for describing the behavior of both pure substances and mixtures, particularly in the petroleum industry.

$$P = \frac{RT}{V_m - b} - \frac{a(T)}{V(V_m + b) + b(V_m - b)}$$

## Virial Equation of State
A series expansion-based equation of state that accounts for the interactions between molecules in a more general way. The virial equation can be truncated at various orders to increase accuracy. The Virial Equation of State is based on the concept that the properties of a real gas can be described as deviations from those of an ideal gas using a series expansion. 

$$ PV = RT \left( 1 + \frac{B(T)}{V_m} + \frac{C(T)}{V_m^2} + \frac{D(T)}{V_m^3} + \ldots \right) $$

## Soave-Redlich-Kwong Equation
An improvement over the original Redlich-Kwong equation, which includes adjustable parameters for different substances to better predict behavior near the critical point.

$$ P = \frac{RT}{V_m - b} - \frac{a(T)}{V_m(V_m + b) + b(V_m - b)} $$

## Cubic Equations of State
Several of the equantions of state are collectivitely referred to as cubic equations of state. These include, but are not limited to, the Van der Waals, Redlich-Kwong, and Soave-Redlich-Kwong equations. They are called "cubic" because they involve a cubic equation for the volume when solved explicity for $V$.

These equations of state are used in various scenarios, including predicting phase transitions, calculating thermodynamic properties, designing and analyzing chemical processes, and more. The choice of equation of state depends on the nature of the substance being studied and the accuracy required for the specific application.

## Terminology

Constant thermodynamic processes describe changes in a system's state variables (such as temperature, pressure, and volume) under specific conditions. There are four main types of constant thermodynamic processes:

1. **Isothermal Process (Constant Temperature):** In an isothermal process, the temperature of the system remains constant throughout the process. The equation for an isothermal process involving an ideal gas is given by:

   $$PV = \text{constant} $$

   Where $ P $ is pressure and $ V $ is volume. Plots of isothermal processes are called `isotherms`.

2. **Isobaric Process (Constant Pressure):** In an isobaric process, the pressure of the system remains constant while other parameters may change. For an ideal gas, the equation for an isobaric process is:

   $$V/T = \text{constant} $$

   Where $ V $ is volume and $ T $ is temperature. Plots of isobaric processes are called `isobars`.

3. **Isochoric Process (Constant Volume):** In an isochoric process, also known as an isovolumetric or isometric process, the volume of the system remains constant. No work is done by or on the system due to volume change in an isochoric process. There is no simple equation relating pressure and temperature in this case. Plots of isochoric processes have no special name.

4. **Adiabatic Process (No Heat Exchange):** In an adiabatic process, there is no exchange of heat with the surroundings. The equation relating pressure, volume, and temperature is specific to the process and may involve the heat capacity ratio ($ \gamma $) of the gas:

   $$PV^\gamma = \text{constant} $$

   Where $ P $ is pressure, $ V $ is volume, and $ \gamma $ is the heat capacity ratio ($ \gamma = \frac{C_p}{C_v} $, where $ C_p $ is the heat capacity at constant pressure and $ C_v $ is the heat capacity at constant volume). Plots of adiabatic processes are called `adiabats`.

These processes describe specific conditions and constraints on the variables in a thermodynamic system. The equations provided above are simplified forms for ideal gases; for real gases or more complex scenarios, the equations might involve additional terms or considerations.

## Understanding and Working with Multivariable Systems in Thermodynamics

In the realm of thermodynamics, the concept of partial derivatives serves as a crucial analytical tool for comprehending the intricate relationships between different thermodynamic properties. Thermodynamic systems often involve multiple variables, and partial derivatives offer a way to dissect how changes in one variable influence other variables while keeping certain factors constant.

In thermodynamics, systems are frequently characterized by a multitude of variables, including pressure ($P$), volume ($V$), temperature ($T$) and others. Consider the deal gas rexpressed as to calculate volume,

 $$V_m = \frac{RT}{P} = R\cdot \bigg(\frac{T}{P}\bigg) = R\cdot f(P, T)$$

 These variables, $T$ and $P$, are interconnected through their equation of state. When changing the system's Temperature, it's pressure and volume will be effected if we don't control for it. It becomes pivotal to gauge the impact of changes in one variable on others. Herein lies the significance of partial derivatives.

### Partial Derivatives

Partial derivatives quantify how a function changes with respect to a single independent variable while keeping other independent variables constant. The symbol $\partial$ is used to indicate a partial derivative, it is no different for th `d` notation for derivatives. For a function $V = f(P, T)$, the partial derivative with respect to $ P $ is denoted as $ \bigg(\frac{\partial V}{\partial P}\bigg)_T $, representing the rate of change of $ V $ with $ P $, while $ T $ is held constant (isothermal). Similar partial derivatives can be calculated for other variables.

### Application of Partial Derivatives

Partial derivatives are instrumental in the thermodynamic toolkit for comprehending how functions, which are reliant on multiple variables (e.g., $f(P, V, T)$), alter with respect to specific variables while the rest are held constant. Examples include:

- $\big(\frac{\partial f}{\partial P}\big)_{V,T}$ denotes the alteration in the function $f$ attributed to changes in pressure, while volume and temperature remain fixed.
- $\big(\frac{\partial f}{\partial V}\big)_{P,T}$ illustrates how $f$ transforms concerning volume, while pressure and temperature are maintained constant.
- $\big(\frac{\partial f}{\partial T}\big)_{P,V}$ represents how $f$ changes due to temperature fluctuations, while pressure and volume remain unaltered.

As long as all variables except for one are held constant in partial derivative then the partial derivative is evaluated in the exact same manner as you are used to doing.

## Partial Derivatives and the Total Differential

In the realm of thermodynamics, **partial derivatives** and the **total differential** are foundational concepts that facilitate the analysis of how changes in multiple variables influence the behavior of a system. These concepts are fundamental for understanding the relationships between various thermodynamic properties and the processes they undergo.

### Total Differential

The total differential accounts for changes in all relevant independent variables simultaneously for the entire equation of state. It provides a comprehensive view of how a function changes due to infinitesimal changes in *all* variables. The total differential of $ V $, denoted as $ dV $, is expressed as the sum of partial derivatives multiplied by the corresponding differentials:

$$ dV = \bigg(\frac{\partial V}{\partial P}\bigg)_T dP + \bigg(\frac{\partial V}{\partial T}\bigg)_P dT $$

Here, $ dP $ and $ dT $ represent infinitesimal changes in $ P $ and $ T $. The total differential considers the collective impact of changes in all variables on the function.

### Practical Significance

We can measure the quantities in the lab, hence, partial derivatives hold pivotal roles in evaluating processes encompassing heat exchange, work accomplished, and alterations in thermodynamic properties during various procedures (such as isothermal and isobaric processes). They facilitate the calculation of shifts in entropy, internal energy, and enthalpy under differing conditions. Moreover, these derivatives are instrumental in establishing Maxwell's relations and ascertaining the thermodynamic connections between diverse properties.

In summary, partial derivatives within the realm of thermodynamics are potent tools that illuminate the repercussions of modifying one variable on others, while simultaneously considering the intricate connections among multiple variables. By virtue of their functionality, they constitute the foundation for dissecting and quantifying the multifaceted interplays within thermodynamic systems. This enables us to predict, interpret, and delve into the behavior of energy and matter under a diverse array of circumstances.

Partial derivatives and the total differential are closely intertwined. The total differential offers a way to calculate how a function changes when all variables change together, while partial derivatives allow us to isolate the effect of changing one variable while keeping others constant. Relationships involving partial derivatives can often be integrated to obtain the total differential.

# Some Important Partial Derivatives in Thermodyanmics

Also these are just for the three state variables of $P$,$V_m$ and $T$. All of theses partials can be measured experimentally, hence they are special since we can verify that they true with empirical evidence. 

$$ \left( \frac{\partial V_m}{\partial T} \right)_P \quad \left( \frac{\partial V_m}{\partial P} \right)_T  \quad \left( \frac{\partial P}{\partial V_m} \right)_T \quad \left( \frac{\partial P}{\partial T} \right)_{V_m} \quad \left( \frac{\partial T}{\partial V_m} \right)_P  \quad \left( \frac{\partial T}{\partial P} \right)_{V_m}$$

However, two of them are far more special than the other four. After a little bit of math we can figure out that there are two independent partial derivatives and that as long as you can measure them, you can calculate the other four partial derivatives. The two independent partials are given special names and symbols and are outlined below.

### Thermal Expansivity

The **thermal expansivity** (also known as coefficient of volume expansion) of a material is denoted by the Greek letter $ \alpha $ (alpha). It quantifies how the volume of a substance changes with temperature while pressure is held constant. Mathematically, it is defined as the fractional change in volume per unit change in temperature:

$$ \alpha = \frac{1}{V_m} \left( \frac{\partial V_m}{\partial T} \right)_P $$

Here:
- $ V_m $ represents the molar volume of the material.
- $ T $ is the temperature.
- $ \left( \frac{\partial V_m}{\partial T} \right)_P $ denotes the partial derivative of volume with respect to temperature, at constant pressure.

In words, the thermal expansivity tells us how much a material's volume will change for a given temperature change while keeping pressure constant.

### Isothermal Compressibility

The **isothermal compressibility** of a substance is denoted by $ \beta $ (beta). It measures how the volume of a material changes with a change in pressure while the temperature is held constant. Mathematically, it is defined as the fractional change in volume per unit change in pressure:

$$ \beta = -\frac{1}{V_m} \left( \frac{\partial V_m}{\partial P} \right)_T $$

Here:
- $ V_m $ represents the molar volume of the material.
- $ P $ is the pressure.
- $ \left( \frac{\partial V_m}{\partial P} \right)_T $ denotes the partial derivative of volume with respect to pressure, at constant temperature.

In simple terms, the isothermal compressibility indicates how much a material's volume changes in response to a change in pressure while keeping temperature constant.
